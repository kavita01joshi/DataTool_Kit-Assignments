4- Explain the use of the groupby() method in Pandas?
* The groupby() method in Pandas is powerful tool for grouping and aggregating data. it work in three step, "It allows you to split your data into groups based on some criteria, apply a function to each group independently, and then combine the result".

 ex:

```python
import pandas as pd
data = {
    'Department': ['HR', 'IT', 'Finance', 'HR', 'IT', 'Finance'],
    'Employee': ['Alice', 'Kavita', 'test1', 'Eve','Qatesting'],
    'Salary': [50000, 60000, 55000, 52000, 62000, 58000]
}
df = pd.DataFrame(data)

# Group by Department and calculate average salary
grouped = df.groupby('Department')['Salary'].mean()
print(grouped)
```

15. How does NumPy handle multidimensional arrays?

NumPy handles multidimensional arrays with remarkable efficiency and flexibility through its core data structure: the ndarray (n-dimensional array).

how it works-
**ndarray Structure** -
* Every NumPy array is an instance of numpy.ndarray.

* It can have any number of dimensions—1D, 2D, 3D, or more.

* The shape of the array is defined by a tuple: e.g., (3, 4, 2) means 3 blocks, each with 4 rows and 2 columns.

**Efficient Storage & Operations** -
* Arrays are stored in contiguous memory, making operations fast.

* NumPy uses vectorized operations, meaning it applies functions across entire arrays without explicit loops.

**Broadcasting** -
* NumPy can perform operations on arrays of different shapes by automatically expanding dimensions.

* This avoids manual reshaping and makes code cleaner.

**Indexing & Slicing** -
* You can access elements using multi-level indexing: arr[1, 2, 3].

* Supports slicing across dimensions: arr[:, 0, :] selects all rows, first column, all depth.

**Reshaping & Transposing** -
* Use reshape() to change dimensions without changing data.

* transpose() or .T flips axes—useful for matrix operations.

**Axis-Based Computation** -
* Many functions like np.sum, np.mean, etc., accept an axis argument.

* This lets you compute across rows, columns, or other dimensions.

17. Explain the difference between apply() and map() in Pandas?
**map() — Element-wise for Series**
* Used only on Series (i.e., a single column).
* Applies a function element by element.
* Ideal for simple transformations or lookups.

**apply() — Flexible and Powerful**
* Works on both Series and DataFrames.
* On a Series: behaves like map(), but more versatile.
* On a DataFrame: applies a function to each row or column.

```python
# 4- How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import seaborn as sns
import matplotlib.pyplot as plt

# Create a sample DataFrame
data = {
    'Math': [88, 92, 80, 89, 100],
    'Science': [90, 85, 88, 95, 92],
    'English': [85, 87, 78, 90, 86]
}
df = pd.DataFrame(data)

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Visualize the correlation matrix using a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
